```python
from statsmodels.tsa.arima.model import ARIMA
from pmdarima import auto_arima

# Automatically determine best p, d, q values
auto_model = auto_arima(y_train, exogenous=exog_train, seasonal=False, trace=True)
p, d, q = auto_model.order
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Train ARIMAX model
model = ARIMA(y_train, order=(p, d, q), exog=exog_train)
fitted_model = model.fit()
print(fitted_model.summary())  # Model Summary
```
